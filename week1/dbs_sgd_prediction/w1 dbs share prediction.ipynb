```python
import pandas as pd
df = pd.read_csv('/content/DBS_SingDollar.csv')
df
```

```python
# Set 'SGD' as Independent Variable (Exchange Rate) and 'DBS' as Dependent Variable (Stock Price)
X = df.loc[:,['SGD']] # Selects all rows and the column 'SGD', store result as DataFrame
Y = df.loc[:,['DBS']] # Selects all rows and the column 'DBS', store result as DataFrame
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Using linear regression for prediction
model = LinearRegression()
model.fit(X,Y)
prediction = model.predict(X)
rmse = mean_squared_error(Y,prediction)**0.5
print(rmse)
```

```python
import joblib # Used to save the trained model into a pickle file
joblib.dump(model,'DBS_SGD_Prediction.pkl')
```

```python
model.coef_
```

- This is the slope (weight) learned by the model

- Since you used one feature (SGD), there is one coefficient

**Interpretation**:

> For every 1 unit increase in SGD, the predicted DBS value decreases by ~50.6 units

```python
model.intercept_
```

- This is the bias term

- Value of DBS when SGD = 0 (theoretical baseline)

The full prediction equation

The model represents:

`DBS = (−50.6009) × SGD + 90.2286`
